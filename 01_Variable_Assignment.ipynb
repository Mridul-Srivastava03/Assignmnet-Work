3. Store the principal amount, rate of interest, and time in different variables and then calculate the Simple Interest for 3 years**.**

```python
p_amount = float(input("Enter the principal amount: "))
r_interest = float(input("Enter the rate of interest: "))
time = float(input("Enter time in years: "))
simple_interest = (p_amount * r_interest * time) / 100
print("Simple Interest: ", simple_interest)
```
